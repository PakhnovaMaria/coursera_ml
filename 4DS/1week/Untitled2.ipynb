```python
file = open('pines.txt', 'r')
```

```python
X = []
Y = []
for line in file:
    #print(line[0])
    if line[0] == 's':
        print(line)
    else:
        a,b = line.split()
        X.append(float(a))
        Y.append(float(b))
```

```python
X
```

```python
ret = sts.binned_statistic_2d(X, Y, X, 'count', bins=[5,5])
```

```python
ret.statistic
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize = (16,9))
plt.plot(X,Y, 'b*')
plt.plot(40*np.ones(200), np.linspace(0,200,200), 'r*')
plt.plot(np.linspace(0,200,200), 40*np.ones(200), 'r*')
plt.show()
```

```python
N = len(X)
N
```

```python
Correct = []
for i in range(5):
    Correct.append(N/25*np.ones(5))
np.array(Correct)
```
