```python
from sklearn import ensemble, model_selection, metrics 

import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_digits
from sklearn.datasets import load_breast_cancer

X_digits = pd.DataFrame(load_digits().data)
y_digits = pd.DataFrame(load_digits().target) 

X_breast_cancer = pd.DataFrame(load_breast_cancer().data)
y_breast_cancer = pd.DataFrame(load_breast_cancer().target)
```

```python
X_digits.head()
#X_breast_cancer.head()
#y_digits.head()
```

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
from sklearn.model_selection import cross_val_score

cif = [BernoulliNB(), MultinomialNB(), GaussianNB()]
for one in cif:
    print(one)
    print('digits', cross_val_score(one, X_digits, y_digits).mean())
    print('breast_cancer', cross_val_score(one, X_breast_cancer, y_breast_cancer).mean())

```
